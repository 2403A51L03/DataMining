```
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
```

```
df=pd.read_csv('/content/titanic.csv')
```

```
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```
df['Age'].fillna(df['Age'].mode(),inplace=True)
```

```
print(df.isnull().sum())
```

```
df['Age']=(df['Age']-df['Age'].min())//(df['Age'].max()-df['Age'].min())
```

```
df['Age']
```

```
scaler = MinMaxScaler()
```

```
columns_to_normalize = ['Age', 'Fare']
```

```
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
```

```
df[columns_to_normalize].head()
```

```
Sscaler = StandardScaler()
```

```
df[columns_to_normalize] = Sscaler.fit_transform(df[columns_to_normalize])
```

```
df[columns_to_normalize].head()
```

```
print("Missing values before imputation:")
print(df.isnull().sum())
```

```
# Handle missing values (impute with mean, median, or mode)
df['Age'].fillna(df['Age'].mean(), inplace=True) # Impute with mean
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True) # Impute with mode
```

```
# Check for missing values after imputation
print("\nMissing values after imputation:")
print(df.isnull().sum())
```

```
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
